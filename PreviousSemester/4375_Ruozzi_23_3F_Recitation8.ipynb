#$\color{blue}{\text{Part 1:}} \text{ Probability}$

We can describe a space of things that happen, denoted $\Omega$.

For instance:
* flipping a coin gives $\Omega=\{H,T\}$
* rolling a six-sided die gives $\Omega=\{1, 2, 3, 4, 5, 6\}$.

Each of those values is an **outcome**.

An **event** is a combination of those outcomes. We may care about rolling a 2 or a 3, the event $A=\{2,3\}\subseteq \Omega$.

Each individual outcome is an event. All combinations of outcomes are also events. The space of all outcomes is the power space $2^\Omega$. This can be confusing. Don't let it be.

Probability is a function. It maps events from $\Omega$ to $[0,1]$.
$$P: 2^\Omega \longrightarrow [0,1]$$

We know a few basic things about probabilities:
* Every probability is non-negative
* Probabilities of **outcomes** are additive
* Probabilities of **outcomes** sum to 1

That is
* $p(\omega) \in [0,1]$
* $\sum_\omega p(\omega) = 1$

We can talk about combinations of events. If $A,B \subseteq\Omega$, we can talk about $P(A \cap B)$, the probability of both of those events. We can also talk about $P(A|B)$, the conditional probability of $A$ when we know $B$ has occurred. They're related by Bayes' Rule
$$
P(A|B)=\frac{P(A \cap B)}{P(B)}
$$



We can also talk about independence: if $A \perp B$, then it doesn't matter what we know about $P(A)$ for $P(B)$
$$
P(A \cap B) = P(A)P(B)
$$

If $A$ and $B$ are independent, what can we say about $P(A|B)$? How?

#$\color{blue}{\text{Part 2:}} \text{ Random Variables}$

We discussed events, which are elements of the state space $\Omega$. Those are weird lumpy things with texture and they may smell funny. We'd rather deal with discrete numbers (bonus: sterile!).

We can use a discrete random variable to move from gross states to pleasant discrete numbers. A random variable is a function. It maps events to values.

Now, we can define a random variable (function) $X$. Let's play with the state space $\Omega = \{H, T\}$.

$$
X(H)=1\\
X(T)=0
$$

Now we can talk about the probability of those random variables. Just like we said

$$P(H)$$

"the probability of heads," we can say

$$P\big(X(H)=1\big)$$

"the probability of heads" or even just

$$P(X=1)$$

"the probability of heads." So,

$$P(H)=P\big(X(H)=1\big)=P(X=1)=0.5$$

The probability of heads is $0.5$.

It feels weird. It may feel messy. It is far more cleaner this way.

We can also have continuous random variables. That way lies madness.

Now we can more abstractly refer to $P(X=x)$ for instantiations, $x$, of $X$. We can say things like
$$1 = \sum_x P(X=x)$$

Just like before, we can talk about probabilities of combinations of things. With random variables $X$ and $Y$, we can have $P(X=y, Y=y)$, the joint probability, just as before.

Confusingly (but more succinctly), we can drop the random variable and refer to just $P(x,y)$. This is the same as $P(X=x, Y=y)$.

It is subtle, but important, to not confuse events (like $A$ and $B$ from before) with random variables (like $X$ and $Y$).

ALMOST THERE

We have $X$ and $Y$. They take on values. We can define sums over them. We can (FINALLY) describe marginal probability. For discrete random variables,

$$
P(A=a)=\sum_bP(A=a, B=b)
$$

Here's a hypothetical joint probability table:

A \ B  | $b_1$ |  $b_2$ | $b_3$ | $P(A)$
-------|-------|--------|-------|--------
$a_1$  | 0.1   | 0.05   | 0.05  | 0.2
$a_2$  | 0.1   | 0.2    | 0.1   | 0.4
$a_3$  | 0.1   | 0.1    | 0.2   | 0.4
$P(B)$ | 0.3   | 0.35   | 0.35  | 1.0

We can find $P(A=a_1)$

$$
\begin{align}
P(A=a_1)
&= \sum_i P(A=a_1, B=b_i) \\
&= P(A=a_1, B=b_1) + P(A=a_1, B=b_2) + P(A=a_1, B=b_3) \\
&= 0.1 + 0.05 + 0.05 \\
&= 0.2
\end{align}
$$

#$\color{blue}{\text{Part 3:}} \text{ Expectation}$


Expectation is what we think we're gonna get. <!-- It's premeditated resentment -->

If there's a \\$1 raffle with a 1% chance of winning \\$50, and a 99% chance of getting nothing, we can expect to lose a \\$0.50 when we play.

That's expectation!

$$
\mathbb{E}[Z] = \sum_{z\in Z} zP(Z=z)
$$

Let $Z$ be the results of the raffle.

$$
\begin{align}
\mathbb{E}[Z]
&= \sum_{z\in Z} zP(Z=z) \\
&= \$50 \times 0.01 + \$0 \times 0.99 \\
&= \$0.50
\end{align}
$$

Should we play? No. We have to spend a dollar to do it.


Expectation is linear:

$$
\mathbb{E}[X+Y] = \mathbb{E}[X] + \mathbb{E}[Y]
$$

We can show this:

$$
\begin{align}
\mathbb{E}[X+Y]
&= \sum_{x,y}(x+y)P(X=x \text{ and } Y=y) \\
&= \sum_{x,y}xP(X=x \text{ and } Y=y) + \sum_{x,y}yP(X=x \text{ and } Y=y) \\
&= \sum_{x}\sum_{y}xP(X=x \text{ and } Y=y) + \sum_{y}\sum_{x}yP(X=x \text{ and } Y=y) \\
&= \sum_{x}x\sum_{y}P(X=x \text{ and } Y=y) + \sum_{y}y\sum_{x}P(X=x \text{ and } Y=y) \\
&= \sum_{x}xP(X=x) + \sum_{y}yP(Y=y) \\
\end{align}
$$


Expectation is also preserved when scaled. For some scalar $a$:

$$
\mathbb{E}[aX] = a\mathbb{E}[X]
$$

This can be shown as above.

In general, the expectation operator is "affine":

$$
\mathbb{E}[aX+bY] = a\mathbb{E}[X] + b\mathbb{E}[Y]
$$

The expectation of a product of two random variables is not, generally, the product of the expectation of each.

$$ \mathbb{E}[XY] \not= \mathbb{E}[X] \times \mathbb{E}[Y] $$

The exception is when $X$ and $Y$ are independent. **This is a key point**

Generally, though,

$$ \mathbb{E}[XY] = \mathbb{E}[X] \times \mathbb{E}[Y] + \text{Cov}(X, Y)$$


We have a word for the expected value of a random variable: the mean! We refer to is as $\mu=\mathbb{E}[Z]$

Now we can compute the variance of a random variable. It's simply how much we **expect** and given realization to differ from its **expected** value, squared.

<!-- Showing ALLLLLLL the steps -->
$$
\begin{align}
\text{Var}(Z)
&=\mathbb{E}\left[\Big(Z-\mathbb{E}[Z]\Big)^2\right] \\
&=\mathbb{E}\left[Z^2 - 2Z\cdot\mathbb{E}[Z] + \big(\mathbb{E}[Z]\big)^2\right] \\
&=\mathbb{E}\left[Z^2 - 2Z\mu + \mu^2\right] \\
&=\mathbb{E}\left[Z^2\right] - \mathbb{E}\left[2Z\mu\right] + \mathbb{E}\left[\mu^2\right] \\
&=\mathbb{E}\left[Z^2\right] - 2\mu\mathbb{E}\left[Z\right] + \mu^2 \\
&=\mathbb{E}\left[Z^2\right] - 2\mu^2 + \mu^2 \\
&=\mathbb{E}\left[Z^2\right] - \mu^2 \\
&=\mathbb{E}\left[Z^2\right] - \mathbb{E}\left[Z\right]^2 \\
\end{align}
$$
